```python
## Defining the player and dealer for simplicity
class player:
    def __init__(self, state):
        self.sum = state[0]
        self.ace = state[1]
        self.dealer_card = state[2]
        self.state = state
        self.lose = False
    def add(self, val):
        ## Change ace if sum exceeds 21
        if self.ace:
            if self.sum + val > 21:
                self.ace = False
                self.sum -= 10
        self.sum += val
        self.state = (self.sum, self.ace, self.dealer_card)
        if self.sum > 21:
            self.lose = True
                
class dealer:
    def __init__(self, card):
        self.card = [11 if card == 1 else card]
        self.lose = False
        self.ace = True if card == 1 else False
    def add(self, card):
        ## Dealer changes from ace to no ace and vice versa
        if self.ace:
            temp = []
            if sum(self.card) + card > 21:
                f = False
                for i in self.card:
                    if i == 11 and not f:
                        f= True
                        self.ace = False
                        temp.append(1)
                    else:
                        temp.append(i)
            else:
                temp = self.card
            self.card = temp
            self.card.append(card)
        else:
            if card == 1:
                if sum(self.card) + 11 <= 21:
                    self.card.append(11)
                else:
                    self.card.append(1)
            else:
                self.card.append(card)
        if sum(self.card) > 21:
            self.lose = True
        
    def move(self):
        if sum(self.card) > 17:
            return False
        else:
            return True
        
    def get_sum(self):
        return sum(self.card)
```

```python
def get_card():
    r = random.randint(1,13)
    return min(r, 10)
    
## Episode Generation for blackjack
def run_episode(pi):
    ## Random Initial States
    
    ## Exploring starts
    p_sum = random.randint(12,21)
    ace = bool(random.randint(0,1))
    d_card1 = get_card()
    d_card2 = get_card()
    
    p = player((p_sum, ace, d_card1))
    d = dealer(d_card1)
    d.add(d_card2)
    
    state = []
    a0 = random.randint(0,1)
    state.append((p.state,a0))
    
    if p.sum == 21 and d.get_sum() == 21:
        return 0, state
    
    if p.sum == 21:
        d.lose = True
        return 1, state
    
    if a0:
        while not p.lose:
            p.add(get_card())
            if p.lose:
                return -1, state
            a = pi[p.state]
            state.append((p.state,a))
            if not a:
                break
        if p.lose:
            return -1, state
        
    if p.lose:
        return -1, state
    
    while d.move() and not d.lose:
        d.add(get_card())
        
    if d.lose:
        return 1, state
    if abs(21 - p.sum) < abs(21 - d.get_sum()):
        return 1, state
    elif abs(21 - p.sum) > abs(21 - d.get_sum()):
        return -1, state
    else:
        return 0, state
```

```python
## Figure 1

no_of_states = 200
states = []
pi = {}
for player_sum in range(12, 22):
    for ace in range(0,2):
        for opp_card in range(1, 11):
            this_state = (player_sum, bool(ace), opp_card)
            states.append(this_state)
            pi[this_state] = 1
            pi[this_state] = 1
            if player_sum == 20 or player_sum == 21:
                pi[this_state] = 0
                pi[this_state] = 0

def MC_pred(pi, epoch):
    V = {}
    counts = {}
    for i in states:
        V[i] = random.random()
        counts[i] = 0
    episode_count = 0
    while episode_count < epoch:
        G, state = run_episode(pi)
        processed = {}
        for s in state:
            s = s[0]
            if s in processed:
                continue
            processed[s] = True
            counts[s] += 1
            V[s] = V[s] + ((G - V[s])/counts[s])
        episode_count += 1
        if episode_count%1000==0:
            print('\rEpisode {}/{}.'.format(episode_count, epoch), end = "")
    return V
```

```python
def ordinary_sampling_mc(epoch):
    Q = {}
    counts = {}
    pi = {}
    for i in states:
        pi[i] = 1
        if i[0] == 20 or i[0] == 21:
            pi[i] = 0

    for i in states:
        for j in [0,1]:
            Q[(i,j)] = -1
            counts[(i,j)] = 0
        
    episode_count = 0
    val = np.zeros(shape=(epoch))
    while episode_count < epoch:
        G, state = run_episode(pi)
        processed = {}
        W = 1
        for s in state:
            st, at = s[0], s[1]
            if (st, at) in processed:
                continue
            processed[(st, at)] = True
            counts[(st, at)] += 1
            Q[(st, at)] = Q[(st, at)] + ((G*W - Q[(st, at)])/counts[(st, at)])
            W = W*2
        val[episode_count] = max(Q[((13, 1, 1), 0)], Q[((13, 1, 1), 1)])
        episode_count += 1
        if episode_count%1000==0:
            print('\rEpisode {}/{}.'.format(episode_count, epoch), end = "")
    return val
```

```python
def weighted_sampling_mc(epoch):
    Q = {}
    counts = {}
    pi = {}
    for i in states:
        pi[i] = 1
        if i[0] == 20 or i[0] == 21:
            pi[i] = 0

    for i in states:
        for j in [0,1]:
            Q[(i,j)] = -1
            counts[(i,j)] = 0
        
    episode_count = 0
    val = np.zeros(shape=(epoch))
    while episode_count < epoch:
        G, state = run_episode(pi)
        processed = {}
        W = 1
        for s in state:
            st, at = s[0], s[1]
            if (st, at) in processed:
                continue
            processed[(st, at)] = True
            counts[(st, at)] += W
            Q[(st, at)] = Q[(st, at)] + (W/counts[(st, at)])*((G - Q[(st, at)]))
            W = W*2
        val[episode_count] = max(Q[((13, 1, 1), 0)], Q[((13, 1, 1), 1)])
        episode_count += 1
        if episode_count%1000==0:
            print('\rEpisode {}/{}.'.format(episode_count, epoch), end = "")
    return val
    
```

```python
## Episode Generation for Monte Carlo Control

def episode():
    s = 1
    states = []
    rewards = []
    actions = []
    while s!= 0 and s!= 6:
        a = random.choice([-1,1])
        states.append(s)
        actions.append(a)
        s += a
        if s == 0:
            rewards.append(0)
            break
        elif s == 6:
            rewards.append(1)
            break
        else:
            rewards.append(0)
            
    return states, rewards, actions
```

```python
## TD

def td(alpha):
    V_s = []
    V = [0.5]*7
    V[0] = 0
    V[6] = 0
    for e in range(101):
        s = 3
        V_s.append(V.copy())
        while s!= 0 and s!= 6:
            
            ## Update after every step
            a = random.choice([-1,1])
            s_dash = s + a
            r = 0
            if s_dash == 6:
                r = 1
            else:
                r = 0
            V[s] = V[s] + (alpha*(r + V[s_dash] - V[s]))
            s = s_dash
    return V_s
```

```python
## Cliff walking

states = []
for i in range(4):
    for j in range(12):
        states.append((i,j))
actions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
terminal = (3, 11)
init = (3, 0)
```

```python
def trans(s, a):
    ## Gives the next state given the current state, action
    x, y = s
    if x==3 and 1 <= y <= 10:
        return init, -100
    else:
        dx, dy = a
        x = x + dx
        y = y + dy
        return (x, y), -1
    
def valid(s, a):
    ## Determines whether the action a is valid for the state s.
    x, y = s
    dx, dy = a
    if not (0<=x+dx<=3) or not(0<= y+dy<= 11):
        return False
    else:
        return True
```

```python
def get_action(Q, e, s):
    ## Epsilon Greedy choice for the current action value function Q, from the state s, with epsilon = e
    a_max = -1
    m = -1e9
    allow = []
    for a in actions:
        if valid(s, a):
            allow.append(a)
            if m < Q[(s,a)]:
                m = Q[(s,a)]
                a_max = a
    
    if random.random() < e:
        return a_max
    else:
        return random.choice(allow)
```

```python
def q_learning(alpha, e):
    ## Q- Learning Init
    Q = {}
    for s in states:
        for a in actions:
            if valid(s, a):
                Q[(s,a)] = random.random()
                if s == terminal:
                    Q[(s, a)] = 0
    rewards = []
    
    for ep in range(500):
        sum_of_rewards = 0
        s = init
        while s != terminal:
            a = get_action(Q, e, s)
            s_dash, r = trans(s, a)
            sum_of_rewards += r
            q_dash = -1e9
            
            ## Pick max Q(S', a) from Q
            for a_dash in actions:
                if valid(s_dash, a_dash):
                    q_dash = max(q_dash, Q[(s_dash, a_dash)])
            Q[(s, a)] = Q[(s, a)] + alpha*(r + q_dash - Q[(s, a)])
            s = s_dash
        rewards.append(sum_of_rewards)
    return rewards
```

```python
def sarsa(alpha, e):
    Q = {}
    for s in states:
        for a in actions:
            if valid(s, a):
                Q[(s,a)] = random.random()
                if s == terminal:
                    Q[(s, a)] = 0
    rewards = []
    for ep in range(500):
        sum_of_rewards = 0
        s = init
        a = get_action(Q, e, s)
        while s != terminal:
            s_dash, r = trans(s, a)
            sum_of_rewards += r
            a_dash = get_action(Q, e, s_dash)
            
            ## Get action a' from Q using the epsilon greedy policy
            Q[(s, a)] = Q[(s, a)] + alpha*(r + Q[(s_dash, a_dash)] - Q[(s, a)])
            s = s_dash
            a = a_dash
        rewards.append(sum_of_rewards)
    return rewards
```
